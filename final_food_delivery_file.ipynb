```python
import sqlite3

conn = sqlite3.connect("restaurants.db")

with open("restaurants.sql", "r") as f:
    sql_script = f.read()

conn.executescript(sql_script)
```

```python
restaurants = pd.read_sql("SELECT * FROM restaurants",conn)
restaurants.head()
```

```python
merged_df = orders.merge(users,on = "user_id", how = "left")
```

```python
final_df = merged_df.merge(restaurants,on = "restaurant_id",how = "left")
```

```python
final_df
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df.to_csv("final_food_delivery_dataset.csv",index = False)
```

```python
final_df['order_date'] = pd.to_datetime(final_df['order_date'])
```

```python
final_df['quarter'] = final_df['order_date'].dt.to_period("Q")
```

## Q1: City with highest revenue from Gold members

```python
final_df[final_df['membership'] == 'Gold'] \
    .groupby('city')['total_amount'] \
    .sum() \
    .sort_values(ascending=False)
```

## Q2: Cuisine with highest average order value

```python
final_df.groupby('cuisine')['total_amount'] \
    .mean() \
    .sort_values(ascending=False)
```

## Q3: Users with total spend > ₹1000

```python
user_spend = final_df.groupby('user_id')['total_amount'].sum()

count = user_spend[user_spend > 1000].nunique()
count
```

## Q4: Rating range with highest revenue

```python
bins = [3.0, 3.5, 4.0, 4.5, 5.0]
labels = ['3.0–3.5','3.6–4.0','4.1–4.5','4.6–5.0']

final_df['rating_range'] = pd.cut(final_df['rating'], bins=bins, labels=labels)

final_df.groupby('rating_range')['total_amount'] \
    .sum() \
    .sort_values(ascending=False)
```

## Q5: Gold members – highest average order value city

```python
final_df[final_df['membership']=='Gold'] \
    .groupby('city')['total_amount'] \
    .mean() \
    .sort_values(ascending=False)
```

## Q6: Cuisine with few restaurants but high revenue

```python
final_df.groupby('cuisine').agg(
    revenue=('total_amount','sum'),
    restaurant_count=('restaurant_id','nunique')
).sort_values('restaurant_count')
```

## Q7: % orders by Gold members

```python
gold_orders = final_df[final_df['membership']=='Gold'].shape[0]
total_orders = final_df.shape[0]

percentage = round((gold_orders / total_orders) * 100)
percentage
```

## Q8: Restaurant with high AOV but < 20 orders

```python
restaurant_stats = final_df.groupby('restaurant_name_x').agg(
    avg_order_value=('total_amount','mean'),
    order_count=('order_id','count')
)

restaurant_stats[
    restaurant_stats['order_count'] < 20
].sort_values('avg_order_value', ascending=False)
```

## Q9: Highest revenue combination

```python
final_df.groupby(['membership','cuisine'])['total_amount'] \
    .sum() \
    .sort_values(ascending=False)
```

## Q10: Highest revenue quarter

```python
final_df.groupby('quarter')['total_amount'] \
    .sum() \
    .sort_values(ascending=False)
```

## Q6. Orders placed in Top Revenue City among Gold members only

```python
top_city = (
    final_df[final_df['membership'] == 'Gold']
    .groupby('city')['total_amount']
    .sum()
    .idxmax()
)
top_city
```

```python
orders_top_city = final_df[
    (final_df['membership'] == 'Gold') &
    (final_df['city'] == top_city)
].shape[0]

orders_top_city
```
